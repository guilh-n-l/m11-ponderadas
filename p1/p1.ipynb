# Pre-processing techniques

As técnicas de pré-processamento em NLP (Processamento de Linguagem Natural) são etapas fundamentais para preparar os dados textuais para análise e modelagem. Essas técnicas ajudam a limpar, normalizar e transformar o texto em uma forma que seja mais adequada para modelos de machine learning e outras tarefas de NLP.

## Tokenization

A tokenização é o processo de dividir um texto em unidades menores chamadas tokens, que podem ser palavras, caracteres ou subpalavras.

```python
tokenize = lambda *s: [nltk.word_tokenize(i) for i in s] if len(s) != 1 else nltk.word_tokenize(s[0])
tokenize.__doc__ = """
Tokeniza múltiplas strings em palavras.

Esta função utiliza a função `word_tokenize` da biblioteca NLTK para dividir várias strings de texto em tokens individuais.

Parâmetros:
    *s (str): Um número variável de strings de texto a serem tokenizadas.

Retorna:
    list: Uma lista de listas, onde cada sublista contém os tokens extraídos de cada string fornecida.
"""

def print_tokens(ls, end="\n"):
    """
    Imprime uma lista de tokens de forma formatada.

    Esta função recebe uma lista de tokens e imprime cada um deles.

    Parâmetros:
        ls (list): A lista de tokens a ser impressa.
        end (str): O valor a ser utilizado no final da string.
    """
    for i, s in enumerate(ls):
        print(s, end=end if i == len(ls) - 1 else " | ")
```

```python
for i in tokenize(*get_quote(10)):
    print_tokens(i, end="\n\n")
```

## Stemming

O stemming é uma técnica de processamento de linguagem natural (NLP) que tem como objetivo reduzir palavras a sua raiz ou forma base.

```python
stem = lambda sl: [nltk.stem.RSLPStemmer().stem(s) for s in sl]
stem.__doc__ = """
Realiza o stemming em uma lista de palavras.

Esta função utiliza o algoritmo `RSLPStemmer` da biblioteca NLTK para aplicar o processo de stemming em uma lista de palavras em português.

Parâmetros:
    sl (list): Uma lista de palavras (ou tokens) a serem processadas e reduzidas à sua forma raiz.

Retorna:
    Uma lista com os radicais dos membros de `sl`.
"""
```

```python
for i in tokenize(*get_quote(10)):
    print_tokens(stem(i), end="\n\n")
```

# Pre-processing techniques from scratch

As técnicas de pré-processamento de texto "do zero" (sem o uso de bibliotecas prontas) envolvem a implementação manual dos passos necessários para preparar os dados textuais.

## Text normalization

O objetivo principal da normalização é transformar o texto bruto em uma forma mais uniforme e padronizada, facilitando a análise e a modelagem. Ao normalizar o texto, você ajuda a reduzir a variabilidade das palavras, o que melhora a eficiência e a performance dos modelos de NLP.

```python
def normalize(sl):
    """
    Normaliza uma lista de token removendo caracteres especiais e convertendo-as para minúsculas.

    Parâmetros:
    sl (list): Lista de tokens a ser normalizada.

    Retorna:
    Lista de tokens normalizados, sem caracteres especiais e em minúsculas, com entradas vazias removidas.
    """
    normalize.REMOVE_TOKENS = "`~!@#$%ˆ&*()-_=+[]{};:'\"\\|,.<>/?¿"
    n_sl = []
    for s in sl:
        for tok in normalize.REMOVE_TOKENS:
            s = s.lower().replace(tok, "")
        n_sl.append(s)
    return [i for i in n_sl if i != ""]
```

```python
for i in tokenize(*get_quote(10)):
    print_tokens(normalize(i), end="\n\n")
```

# Builiding pipeline

O pipeline será construído combinando as técnicas previamente explicadas em um único fluxo de trabalho. Esse processo integrará as etapas de pré-processamento.

```python
def pipeline(st):
    """
    Processa uma string de texto, realizando tokenização, normalização e stemming.

    Parâmetros:
    st (str): A string de texto a ser processada.

    Retorna:
    Uma tupla contendo dois elementos:
        - (0) Uma lista de tokens normalizados.
        - (1) Uma lista de tokens com o stemming aplicado.
    """
    tokens = normalize(tokenize(st))
    return tokens, stem(tokens)
```

```python
for i in get_quote(10):
    print(f"=== str ===\n{i.replace(" ",  "\\s").replace("\n",  "\\n")}")
    res, stems = pipeline(i)
    print("--- tokens ---")
    print_tokens(res)
    print("--- stem ---")
    print_tokens(stems, end="\n=== end ===\n\n")
```
